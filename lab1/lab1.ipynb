```python
t_span = np.linspace(0,1,500)
x0 = [0,1]
```

```python
def plot_solution(solution,title = None):
    for i,xn in enumerate(solution.y):
        plt.plot(solution.t, xn,label=f"x{i+1}(t)")
    plt.xlabel("t")
    plt.ylabel("x(t)")
    if title:
        plt.title(title)
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
# x'(t) = x - 100
# x(0) = 0
# t = [0, 1]

def model_1(t, y):
    return [y[0] - 100]

solution1 = solve_ivp(
    model_1,
    [t_span[0], t_span[-1]],
    [x0[0]],
    t_eval=t_span
)
plot_solution(solution1)


```

```python
def model_2(t, y):
    x1, x2 = y
    return [x1 - 100, x2 - 200]


solution2 = solve_ivp(
    model_2,
    [t_span[0], t_span[-1]],
    x0,
    t_eval=t_span
)

plot_solution(solution2)
```

```python
# x1(t) = x1 - 100 + u
# x2(t) = x2 - 200 + u
# x1(t0) = 0
# x2(t0) = 1
# u = 50
# t = [0,1]
def model_3(t, y, u):
    x1, x2 = y
    return [x1 - 100 + u, x2 - 200 + u]

u = 50

solution3 = solve_ivp(
    model_3,
    [t_span[0], t_span[-1]],
    x0,
    t_eval=t_span,
    args=(u,)
)

plot_solution(solution3,title=f"Output for {u=}")
```
